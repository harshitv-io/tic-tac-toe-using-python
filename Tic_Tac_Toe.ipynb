```python
from IPython.display import clear_output
import random

#DISPLAY BOARD
def display_board(board):
  clear_output()

  print(board[1]+' | '+board[2]+' | '+board[3])
  print('---------')
  print(board[4]+' | '+board[5]+' | '+board[6])
  print('---------')
  print(board[7]+' | '+board[8]+' | '+board[9])

#PLAYER INPUT
def player_input():
  marker = ''
  while not (marker=='X' or marker=='O'):
    marker = input('\nPlayer 1, Choose X or O :').upper()

  if marker=='X':
    return('X','O')
  else:
    return('O','X')

#MARK X O IN BOARD
def mark_in_board(board,marker,position):
  board[position] = marker

#WIN CHECK
def win_check(board,mark):
  return(
      #ROWS
      (board[1] == mark and board[2] == mark and board[3] == mark) or
      (board[4] == mark and board[5] == mark and board[6] == mark) or
      (board[7] == mark and board[8] == mark and board[9] == mark) or
      #COLUMNS 
      (board[1] == mark and board[4] == mark and board[7] == mark) or
      (board[2] == mark and board[5] == mark and board[8] == mark) or
      (board[3] == mark and board[6] == mark and board[9] == mark) or
      #DIAGONAL
      (board[1] == mark and board[5] == mark and board[9] == mark) or
      (board[7] == mark and board[5] == mark and board[3] == mark)
  )

#COMPUTER CHOOSES WHICH PLAYER PLAYS FIRST
def player_choose():
  if random.randint(0,1) == 0:
    return "Player 1"
  else:
    return "Player 2"


#CHECK FOR SPACE
def  space_check(board,position):
  return board[position] == ' '

#CHECK FOR BOARD FULL
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True

#PLAYER INPUT AND CHECK IF BOARD IS FULL
def player_choice1(board):
  position = 0

  while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
    position = int(input("\nPlayer 1, Choose a position from 1-9 : "))
  return position

def player_choice2(board):
  position = 0

  while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
    position = int(input("\nPlayer 2, Choose a position from 1-9 : "))
  return position

#REPLAY MATCH
def replay():
    replay= None
    while not (replay=='Y'or replay=='N'):
      try:
        replay = input("Do you want to play again? (Y/N): ").upper()
      except:
        print('\nInvalid Input')
    return replay == "Y"
    if replay == False:
      print("\n-----Game End-----")

#JOINING ALL

print('-----Welcome to TicTacToe------\n')
while True:

  the_board = [' ']*10

  player1_marker,player2_marker = player_input()

  turn = player_choose()

  print(f'{turn} will play first.')

  play_game = input('\nAre you ready to play? Enter Yes or No: ')

  if play_game.lower()[0]=='y':
    game_on = True
  else:
    game_on = False

  while game_on:
    if turn == 'Player 1':
      display_board(the_board)
      position = player_choice1(the_board)
      mark_in_board(the_board, player1_marker, position)
    
      if win_check(the_board,player1_marker):
        display_board(the_board)
        print("\nPlayer 1 has Won\n")
        game_on = False
      else:
        if full_board_check(the_board):
          display_board(the_board)
          print("\nIt's a Tie\n")
          break
        else:
          turn = 'Player 2'
    else:
      display_board(the_board)
      position = player_choice2(the_board)
      mark_in_board(the_board, player2_marker, position)
    
      if win_check(the_board,player2_marker):
        display_board(the_board)
        print("\nPlayer 2 has Won\n")
        game_on = False
      else:
        if full_board_check(the_board):
          display_board(the_board)
          print("It's a Tie\n")
          break
        else:
          turn = 'Player 1'
  if not replay():
    break
```
